```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe.
df = pd.read_csv("influencer_marketing_roi_dataset.csv")

# Group by platform and calculate the average of key metrics.
platform_metrics = df.groupby('platform')[['engagements', 'estimated_reach', 'product_sales']].mean().reset_index()

print("Platformlara g√∂re ortalama kampanya metrikleri:")
print(platform_metrics)

# Create a bar chart for average product sales by platform.
plt.figure(figsize=(10, 6))
sns.barplot(x='platform', y='product_sales', data=platform_metrics.sort_values(by='product_sales', ascending=False), palette='viridis')
plt.title('Platformlara G√∂re Ortalama √úr√ºn Satƒ±≈üƒ±')
plt.xlabel('Platform')
plt.ylabel('Ortalama √úr√ºn Satƒ±≈üƒ±')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('avg_product_sales_by_platform.png')

# Create a bar chart for average engagements by platform.
plt.figure(figsize=(10, 6))
sns.barplot(x='platform', y='engagements', data=platform_metrics.sort_values(by='engagements', ascending=False), palette='magma')
plt.title('Platformlara G√∂re Ortalama Etkile≈üim')
plt.xlabel('Platform')
plt.ylabel('Ortalama Etkile≈üim')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('avg_engagements_by_platform.png')

# Create a bar chart for average estimated reach by platform.
plt.figure(figsize=(10, 6))
sns.barplot(x='platform', y='estimated_reach', data=platform_metrics.sort_values(by='estimated_reach', ascending=False), palette='cividis')
plt.title('Platformlara G√∂re Ortalama Tahmini Eri≈üim')
plt.xlabel('Platform')
plt.ylabel('Ortalama Tahmini Eri≈üim')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('avg_reach_by_platform.png')
```
